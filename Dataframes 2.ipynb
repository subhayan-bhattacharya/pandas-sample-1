# Filtering a data frame based on multiple condition

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"
df[mask1 & mask2]
```

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"
df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"
df[(mask1 & mask2) | mask3]
```

# Is in method , when we have a multiple set of values against whom we have to check

```python
df[~df["Senior Management"]] # This is a way to negate a condition in pandas. So give me all values for which the Senior Management
# is false
```

```python
df[df["Team"].isin(["Legal", "Sales", "Product"])]
```

```python
df[df["Team"].isnull()].info() # Extract the rows from the dataframe where the Team value is None
```

```python
df[df["Team"].notnull()]
```

```python
df[df["Salary"].between(60_000, 70_000)]
```

```python
df[df["Last Login Time"].between("07:00AM", "09:00AM")]
```

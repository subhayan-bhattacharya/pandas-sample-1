```python
nba.dropna() # Drops any rows inside the dataframe even where there is one or more null values
```

```python
nba.dropna(how='all') # Drops the row if all the columns have null . Does not do it inplace, there is an inplace param 
```

```python
nba.dropna(how='all', inplace=True) # Only removes the last row which has all Null values and i also provide inplace to modify
# permanently
nba.dropna(axis = 1) # Removes the columns which have Null values inside them instead of removing rows which have null values
```

```python
nba.dropna(subset=["Salary"]) # Drop rows which have a null value in the Salary column
```
